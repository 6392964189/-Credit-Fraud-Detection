```python
# Plot the distribution of loan amount
plt.figure(figsize=(10, 6))
sns.histplot(df['loan_amnt'], bins=50, kde=True)
plt.title("Distribution of Loan Amount")
plt.xlabel("Loan Amount")
plt.ylabel("Count")
plt.show()
```

```python
# Plot the relationship between loan amount and interest rate
plt.figure(figsize=(10, 8))
sns.scatterplot(x='loan_amnt', y='int_rate', data=df)
plt.title("Relationship between Loan Amount and Interest Rate")
plt.xlabel("Loan Amount")
plt.ylabel("Interest Rate")
plt.show()
```

```python
# Plot the distribution of interest rate
plt.figure(figsize=(10, 6))
sns.histplot(df['int_rate'], bins=50, kde=True)
plt.title("Distribution of Interest Rate")
plt.xlabel("Interest Rate")
plt.ylabel("Count")
plt.show()
```

```python


# Distribution of loan_status
sns.countplot(x='loan_status', data=df)
plt.title('Distribution of Loan Status')
plt.show()

# Distribution of term
sns.countplot(x='term', data=df)
plt.title('Distribution of Term')
plt.show()

# Distribution of grade
sns.countplot(x='grade', data=df)
plt.title('Distribution of Grade')
plt.show()

# Distribution of home_ownership
sns.countplot(x='home_ownership', data=df)
plt.title('Distribution of Home Ownership')
plt.show()

# Distribution of verification_status
sns.countplot(x='verification_status', data=df)
plt.title('Distribution of Verification Status')
plt.show()

# Distribution of purpose
sns.countplot(x='purpose', data=df)
plt.title('Distribution of Purpose')
plt.xticks(rotation=90)
plt.show()
 
```
